```python
import yfinance as yf
import matplotlib.pyplot as plt
from datetime import datetime

# get data from the start of the year till now
now = datetime.now()
start_date = datetime(now.year, 1, 1)
end_date = datetime(now.year, now.month, now.day)

# download the stock data
nvda_data = yf.download('NVDA', start=start_date, end=end_date)
tesla_data = yf.download('TSLA', start=start_date, end=end_date)

# plot the Close price data
plt.figure(figsize=(14, 7))
plt.plot(nvda_data.index, nvda_data['Close'], label='NVDA')
plt.plot(tesla_data.index, tesla_data['Close'], label='TESLA')

plt.title('NVDA vs TESLA Stock Price Year to Date')
plt.xlabel('Date')
plt.ylabel('Close Price ($)')
plt.legend()
plt.grid(True)
plt.show()
```
